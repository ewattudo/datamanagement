# PS4 - Empirical Ananlysis of Hunter and Robinsons' (2018) C*hocolate Cities* Map

***It’s all the same.***

***It’s all the same*** from New York’s skyline to the rolling cornfields of Indiana.

***It’s all the same*** from Miami’s beaches to the snow-capped mountains of Colorado.

***It’s all the same.*** More specifically, it’s all the South.

Malcolm X made this point in his Ballot or the Bullet speech-a speech he gave in 1964 after his bitter and public divorce from Elijah Muhammad and the Nation of Islam. While the speech was also delivered in Cleveland that same month in 1964, Hunter and Robinson use the Motown backdrop of King Solomon Baptist Church as inspiration for their assertion that, as Malcolm X averred, “As long as you south of the Canadian border, you South.” According to sociologists and anthropologists Hunter and Robinson in their 2018 book, anything below Canada-for Black people in America-is the South.

**Down South, Deep South, Up South, Mid South, Out South, and West South**

They carry this rhetorical tool throughout their work, arguing that the United States consists of Down South, Deep South, Up South, Mid South, Out South, and West South. Their maps (provided here and recreated below) suggest that Out South includes Northern California, Alaska, and the northwestern states from Washington to the easternmost borders of Montana and Wyoming, comprising 3.14% of Black residents in 2010.
West South is its southern neighbor, including Southern California, Colorado, and the entire state of Texas. It borders both the smallest and largest of Hunter and Robinson’s regions. The Deep South, which made up 15.29% of the Black population in 2010, includes Arkansas, Tennessee, Louisiana, Mississippi, and Alabama. Above it is the Mid South, with 25.13% of the population, spanning from the Dakotas south to Oklahoma and stretching to western Pennsylvania and West Virginia.
Then there are the Atlantic “Southern” states. Up South includes the smallest states but the highest concentration of Black residents (29.12%), stretching from Washington, D.C. suburbs in West Virginia to Northern Maine, cutting through Pennsylvania along the way. What remains is Down South, with 16.57% of Black residents, containing the southern part of Virginia to Southern Florida—excluding the Florida Panhandle, which is categorized under Deep South.

The idea that “it’s all the same” is rhetorical, not empirical. That is the question posed by this project. But this is not just an academic exercise. First, there is a dearth of literature that uses quantitative methods to interrogate the relationship between race and geography (Perry, 2014). Second, as Phillips and Pittman (2009) argue, community development is both a noun (the developed space) and a verb (the act of developing).

This work is designed to help fill that gap by testing whether these regions differ meaningfully-and to provide data that can determine where and how the Black community is developing or failing to do so. In this project, for example, I test the relationship between education, poverty, household income, and population density. It’s not enough to identify that differences exist-but also whether those differences are positive or inconsequential. I am not interested in merely inventorying despair or racism, but rather in contributing to community development: both the noun, where it has already emerged, and the verb, as a replicable process for others to follow.

**Method**

The unit of analysis in this project is the state. Despite Hunter and Robinson’s use of cities in the title of their publication, the quantitative data they use is largely at the state-level. The city-based analysis they provide is largely qualitative, poststructural (Perry, 2014; Dixon and Jones, 1997), and rooted in critical theory. This allows them to illustrate lived experience through personal and cultural narratives—from familial relations to figures like Ida B. Wells and Marsha P. Johnson, and pop culture references such as comedian Jordan Peele and singer Solange Knowles. While I acknowledge the value of that approach, I contend that too many studies of Black experiences—what Rabaka (2015) describes - “privilege poetics over politics, aesthetics over decolonization, re-Africanization, and radical humanism.”

First, despite Rao’s (2017) caution against dichotomizing U.S. political divisions, I grouped states by party control using National Council of State Legislatures data. Future studies should move beyond red/blue/purple labels and use empirically grounded political typologies. This grouping was chosen based on both popular familiarity and personal experience. Reimer (1977) encourages researchers to use “opportunistic research” like this when it allows them to engage familiar social contexts with special expertise.

Second, I performed Analysis of Variance (ANOVA) on modified versions of Hunter and Robinson’s regions. The main difference between their regions and mine is that I assigned whole states to a region. I did not have access to the detailed or even general data necessary to split states like Pennsylvania, California, or Virginia like they did, it actually looks a bit arbitrary. I used data from 2010, the latest census year referenced in their work.

The ANOVA utilized 2010 population proportions from Wikipedia but the remaining variables were pulled from the 2010 ACS 5-Year Estimates. These included state-level Black population counts, median household income for Black households, number of Black people in poverty, and counts of Black males and females with at least a bachelor’s degree. I also calculated proportions for education and poverty to control for state-level population differences.

**Findings**

There is a statistically significant difference in every variable across the Chocolate Cities regions.

Population Proportions differed significantly by region (F(5,42) = 4.31, p = 0.003), as did Poverty (F(5,42) = 4.54, p = 0.002), and the Proportion of People in Poverty (F(5,42) = 5.88, p < 0.001). Median Black Household Income varied strongly across regions (F(5,42) = 16.39, p < 0.000001), as did the share of the national Black population living in each state in 2010 (F(5,42) = 12.78, p < 0.000001).

Educational attainment also differed: the number of individuals with a bachelor’s or higher (F(5,42) = 3.21, p = 0.015), the proportion with a bachelor’s or higher (F(5,42) = 7.96, p < 0.001), and both the number of Black females (F(5,42) = 3.38, p = 0.012) and males (F(5,42) = 2.99, p = 0.021) with degrees showed statistically significant differences by region.

### Importing Census Data

```python
#I couldn't get this to work until after I downloaded the data and looked at the spreadsheets. It was hard to find the variables I wanted otherwise.
url = "https://api.census.gov/data/2010/acs/acs5"
params = {
    "get": ",".join([
        "NAME",
        "B19013B_001E",# Median Black Household Income
        "B17001B_002E",# Black in Poverty
        "C15002B_006E",# Male Bachelor's or higher
        "C15002B_011E",# Female Bachelor's or higher
        "B02001_003E"# Black or African American alone
    ]),
    "for": "state:*"
}

response = requests.get(url, params=params)

if response.status_code == 200:
    data = response.json()
    census = pd.DataFrame(data[1:], columns=data[0])
    census = census.rename(columns={
        "NAME": "State",
        "B19013B_001E": "HHInc",
        "B17001B_002E": "Pov",
        "C15002B_006E": "Bach_M",
        "C15002B_011E": "Bach_F",
        "B02001_003E": "Pop"
    })

census.head()
```

### Manipulating Census Data

```python
#I need to change HHInc, Pov, Bach_M,Bach_F, Pop into numbers
census['HHInc'] = census['HHInc'].astype(int)
census['Pov'] = census['Pov'].astype(int)
census['Bach_M'] = census['Bach_M'].astype(int)
census['Bach_F'] = census['Bach_F'].astype(int)
census['Pop'] = census['Pop'].astype(int)
census.dtypes
```

```python
#I want to add Bach_M to Bach_F to create a field called HED
# In the future, I want dropout rate and some college.
census['HED'] = census['Bach_M'] + census['Bach_F']
census.head()
```

```python
#I want to create fields Pov/Pop, Bach_M/Pop, Bach_F/Pop and HED/Pop
census['PovPop'] = census['Pov'] / census['Pop']
census['Bach_MPop'] = census['Bach_M'] / census['Pop']
census['Bach_FPop'] = census['Bach_F'] / census['Pop']
census['HEDPop'] = census['HED'] / census['Pop']
census.head()
```

```python
census.dtypes
```

```python
#I want to get rid of state
census = census.drop(columns=['state'])
census.head()
```

## Creating a Base Map

Throughout Hunter and Robinson (2018), the authors present the map shown below and reference migration from "down South" to the "mid-South" and "up South." For my analysis, I chose to recreate this map so I could overlay data and test the hypothesis that "it's all South."

You’ll notice that the Hunter and Robinson (2018) map differs from mine in two major ways:

1. Their lines appear to be hand-drawn, almost as if with a marker, while mine are generated using python.

2. They divide large states in ways that support their sociological framework—rather than adhering strictly to political boundaries. They categorize regions based on societal similarities. For example, Pennsylvania is split in two to reflect differing affinities with the Midwest or  Mid South, and northern California is treated as distinct from the southern part of the state, with a closer affinity to the Out South and West South respectively.

### Testing Variance Between Regions and Continuous Data (ANOVA and Tukey)

```python
# Variables to analyze
variables_for_anova_by_region = [
    'Pop', 'Pov', 'PovPop', 'HHInc',
    '2010', 'HED', 'HEDPop', 'Bach_F', 'Bach_M'
]

# Store results
anova_results_by_variable = {}

for demographic_variable in variables_for_anova_by_region:
    # Filter out missing values
    anova_data_by_region = US[[demographic_variable, 'AllSouth']].dropna()

    # If the variable name is not a valid formula name (e.g., starts with number), quote it
    if demographic_variable[0].isdigit():
        formula = f'Q("{demographic_variable}") ~ C(AllSouth)'
    else:
        formula = f'{demographic_variable} ~ C(AllSouth)'

    # Fit OLS model
    model_by_region = ols(formula, data=anova_data_by_region).fit()

    # Run ANOVA
    anova_table_by_region = sm.stats.anova_lm(model_by_region, typ=2)

    # Store table
    anova_results_by_variable[demographic_variable] = anova_table_by_region

# Print results
for variable, table in anova_results_by_variable.items():
    print(f"\nANOVA results for {variable} by Chocolate Cities Region:\n")
    print(table)
```
